```python
import pandas as pd
df2 = pd.read_csv("E:/Jane_Street/train_sliced.csv")
dfresp = df2
df2
```

```python
df2= df2.drop(["resp_1","resp_2","resp_3","resp_4","resp"],axis = 1)
df2
```

```python
selectedFeature = ['date',
 'weight', 'feature_0', 'feature_3',
 'feature_5', 'feature_17',
 'feature_27', 'feature_29',
 'feature_37',
 'feature_39',
 'feature_41',
 'feature_42',
 'feature_43',
 'feature_44',
 'feature_45',
 'feature_51',
 'feature_55',
 'feature_60',
 'feature_62',
 'feature_67',
 'feature_75',
 'feature_77',
 'feature_83',
 'feature_90',
 'feature_95',
 'feature_100',
 'feature_102',
 'feature_103',
 'feature_106',
 'feature_120',
 'feature_121',
 'ts_id']
df2 = df2[selectedFeature]
df2
```

```python
Corelation = df2.corr()
```

```python
import seaborn as sns
sns.heatmap(Corelation)
```

```python
import numpy as np
columns = np.full((Corelation.shape[0]), True, dtype=bool)
for i in range(Corelation.shape[0]):
    for j in range(i+1, Corelation.shape[0]):
        if Corelation.iloc[i,j] >= 0.9:
            if columns[j]:
                columns[j] = False

selected_columns_2 = df2.columns[columns]
df2 = df2[selected_columns_2]
selected_columns_2
```

```python
df2
```

```python
df2 = df2.fillna(df2.mean())
df2
```

```python
dfresp["action"] = ((df2["weight"]*dfresp["resp"])>0).astype('uint8')
```

```python
from sklearn.model_selection import train_test_split
seed = 7
y = dfresp["action"].copy()
X_train, X_test, y_train, y_test = train_test_split(df2, y, test_size=0.33,random_state=seed)
X_test
```

```python
from numpy import loadtxt
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
model = xgb.XGBClassifier(n_estimators=350, 
                      max_depth=20,
                      eta=0.01,
                      missing=None,
                      random_state=42,
                      tree_method='hist',
                      subsample=0.8,
                      colsample_bytree=1)
model.fit(X_train,y_train)
y_pred = model.predict(X_test)

predictions = [round(value) for value in y_pred]
accuracy = accuracy_score(y_test, predictions)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
#accuracy 84.88% by selected features
```
